**List vs Numpy - Time Taken**

```python
from time import process_time
```

```python
#time taken by the list
```

```python
py_list = [i for i in range(10000)]

start_time = process_time()

py_list = [i+5 for i in py_list]

end_time = process_time()

print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array = np_array+5

end_time = process_time()

print(end_time - start_time)
```

**Initial placeholders in numpy arrays**

```python
#creating a numpy array of zeros
x = np.zeros((4,5))
print(x)
```

```python
#creating a numpy array of zeros
y = np.ones((3,3))
print(y)
```

```python
# array of particular one
z = np.full((5,4),5)
print(z)
```

```python
# Creating an Identity Matrix
a = np.eye(4)
print(a)
```

```python
#Create an Numpy array with random values
b = np.random.random((3,4))
print(b)
```

```python
#Create an Numpy array with random intrgers values
b = np.random.randint(10,100,(3,5))
print(b)
```

```python
# array of evenly spaced values
e = np.linspace(10,30,5)
print(e)
```

```python
# array of evenly spaced values --> specifying the step
f = np.arange(10,30,5)
print(f)
```

```python
#convert a list to numpy array
list2 = [1,2,3,4,5]
np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

**Mathematical operations on a np array**

```python
list1 = [1,2,3,4,5]
list2 = [5,6,7,8,9]
print(list1+list2)
```

```python
a = np.random.randint(0,10,(3,3))
print(a)
```

```python
b = np.random.randint(10,20,(3,3))
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
```

```python
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
#reshaping an array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```

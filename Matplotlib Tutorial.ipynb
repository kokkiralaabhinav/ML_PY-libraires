**Matplotlib:**  

* Useful for making Plots





```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy library to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

**Ploting the Data**

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sin value')
plt.title('sine wave')
plt.show()
```

```python
#parabole curve
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'r')
plt.show()
```

```python
#Hyperbola curve
plt.plot(y,x)
plt.show()
```

**Bar plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English','Hindi','Telugu','kannada','Tamil']
people = [100,50,150,40,70]
ax.bar(language, people)
plt.xlabel("language")
plt.ylabel("Number of people")
plt.show()
```

**Pie Chart**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English','Hindi','Telugu','kannada','Tamil']
people = [100,50,150,40,70]
ax.pie(people, labels=language, autopct='%1.1f%%')
plt.show()
```

**Scatter plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

**3D Scatter Plot**

```python
fig2 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
